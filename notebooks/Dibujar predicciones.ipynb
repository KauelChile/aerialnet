```python
def get_img(imgURL, outputDir):
    try:
        imgName = imgURL.split('/')[-1]
        imgPath = os.path.join(outputDir, imgName)

        # check if image was already processed
        if len(glob.glob(imgPath.replace('.jpg', '').replace('.JPG', '')+'*')) > 0:
            print('Image {} was already processed'.format(imgName))
            return None, imgPath, None

        print('Downloading image: {}'.format(imgURL))
        image_bytes = requests.get(imgURL)
        pilImg = Image.open(io.BytesIO(image_bytes.content))

        # save original image
        pilImg.save(imgPath)
    except OSError:
        return None, None, None

    return None, imgPath, imgName

def render(img, boxes, scores, labels, LABELS=None, thickness=1) -> np.ndarray:
    """ Dibuja las marcas en una imagen """

    #img_pil = Image.fromarray(img)
    img_pil = Image.open(img)
    draw = ImageDraw.Draw(img_pil)

    # loop over detections
    for (bbox, score, label) in zip(boxes, scores, labels):
        text = label + ' {}%'.format(int(score*100))
        #Determina el color de la marca en base a la clase
        #color = label_color(label)
        color = (255,255,255)

        x1 = bbox[0]
        y1 = bbox[1]
        x2 = bbox[2]
        y2 = bbox[3]
        
        draw.rectangle([(x1,y1),(x2,y2)], outline=color, width=thickness)
        
        draw.rectangle([(x1,y1-7),(x2,y1)], outline=color, width=6)
        draw.text((x1,y1-16/2-4), text, fill=(0,0,0))
    img_pil.save(img)
    return
```
